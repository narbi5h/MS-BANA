```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df=pd.read_csv('Air Cost Dataset.csv')
df.head()
```

```python
df['aircost'] = df['aircost'].round(2)
df['aircost'] = df['aircost'].astype(int)  # Convert to integer
df['avgfare'] = df['aircost'] / df['guests']  # Create avgfare column
print(df.dtypes) 
```

```python
df_europe= df[(df['region'] == 'Europe')]

plt.figure(figsize=(10,6))
sns.lineplot(data=df_europe, x='sailmonth', y='aircost', hue='type')
```
