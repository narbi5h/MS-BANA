```python
import pandas as pd
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Read the CSV file
df = pd.read_csv('Student_Grades.csv')

# Display the contents of the DataFrame
print("Result of df.head()")
print(df.describe())
```

```python
plt.figure(figsize=(6, 3))
plt.title('Correlation between Math Scores and English Scores')
sns.scatterplot(data=df, x='MathScore', y='EnglishScore')
plt.xlabel('Math Scores')
plt.ylabel('English Scores')
```

```python
plt.figure(figsize=(6, 3))
plt.title('Correlation between Math Scores and Science Scores')
sns.scatterplot(data=df, x='MathScore', y='ScienceScore')
plt.xlabel('Math Scores')
plt.ylabel('Science Scores')
```

```python
plt.figure(figsize=(6, 3))
plt.title('Correlation between English Scores and Science Scores')
sns.scatterplot(data=df, x='EnglishScore', y='ScienceScore')
plt.xlabel('English Scores')
plt.ylabel('Science Scores')
```
