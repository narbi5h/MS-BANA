```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

#os.chdir('c:/Users/gio12/OneDrive/Documents/GitHub/MS-BANA/WEEK 3')
```

```python

  # importing pandas library
df = pd.read_csv('Crime_Data.csv') # here we are importing the csv file
#print(df)
#print(df.head()) # print out the first 5 lines of sample file

```

```python
df = df.applymap(lambda x: x.replace(',', '') if isinstance(x, str) else x)
```

```python
df.skew() # checking the skewness of the data
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
sns.lineplot(data=df, x="Year", y="Property crime total",markers="o",ax=ax1, color="r")
sns.lineplot(data=df, x="Year", y="Violent crime total", markers="o", ax=ax2, color="b")
```
