```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python

# importing pandas library
df = pd.read_csv('DJIA.csv') # here we are importing the csv file
print(df)
#print(df.head()) # print out the first 5 lines of sample file
```

```python
#df=df.dropna(axis=1, how='all',inplace=True)
df = df.dropna(axis=1, how='all')
df = df.dropna(axis=0, how='all')
```

```python
df = df.applymap(lambda x: x.replace(',', '') if isinstance(x, str) else x)
df = df.applymap(lambda y: y.replace('%', '') if isinstance(y, str) else y)
```

```python
df.describe()
```

```python
df['Percentage Change'] = df['Percentage Change'].astype(float)
```

```python
plt.figure(figsize=(13, 7))
xticks = range(0, len(df),30)
plt.xticks(xticks, df['Month'].iloc[xticks], rotation=45, ha='right', fontsize=12)
sns.lineplot(data=df, x="Month", y="Percentage Change")
plt.tight_layout()
```
